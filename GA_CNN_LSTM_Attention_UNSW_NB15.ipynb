```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'  # Suppress CUDA warnings

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             classification_report, confusion_matrix)
from sklearn.utils.class_weight import compute_class_weight
from sklearn.tree import DecisionTreeClassifier
import tensorflow as tf
from tensorflow.keras.layers import (Dense, Dropout, Conv1D, LSTM, MaxPooling1D,
                                     MultiHeadAttention, LayerNormalization, Input)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import time

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)



# 1. Load and preprocess the dataset
def load_data():
    train_data = pd.read_csv('/kaggle/input/unsw-nb15/UNSW_NB15_training-set.csv')
    test_data = pd.read_csv('/kaggle/input/unsw-nb15/UNSW_NB15_testing-set.csv')
    
    data = pd.concat([train_data, test_data], ignore_index=True)
    data = data.drop(['id', 'attack_cat'], axis=1)
    
    categorical_cols = ['proto', 'service', 'state']
    for col in categorical_cols:
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col])
    
    X = data.drop('label', axis=1)
    y = data['label'].values
    
    return X, y

# 2. Genetic Algorithm for feature selection
def genetic_algorithm_feature_selection(X, y, n_generations=10, population_size=20):
    n_features = X.shape[1]
    
    def create_population(size):
        return np.random.randint(2, size=(size, n_features))
    
    def fitness(individual, X, y):
        selected_features = np.where(individual == 1)[0]
        if len(selected_features) == 0:
            return 0.0
        X_selected = X.iloc[:, selected_features]
        X_train, X_val, y_train, y_val = train_test_split(X_selected, y, test_size=0.2, random_state=42)
        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_val = scaler.transform(X_val)
        model = DecisionTreeClassifier(max_depth=5)
        model.fit(X_train, y_train)
        return accuracy_score(y_val, model.predict(X_val))
    
    def crossover(parent1, parent2):
        crossover_point = np.random.randint(2, len(parent1) - 1)
        child = np.concatenate([parent1[:crossover_point], parent2[crossover_point:]])
        return child
    
    def mutate(individual, mutation_rate=0.01):
        for i in range(len(individual)):
            if np.random.random() < mutation_rate:
                individual[i] = 1 - individual[i]
        return individual
    
    population = create_population(population_size)
    best_individual = None
    best_fitness = 0.0
    
    print("Running Genetic Algorithm for feature selection...")
    for generation in range(n_generations):
        fitness_scores = np.array([fitness(ind, X, y) for ind in population])
        if fitness_scores.max() > best_fitness:
            best_fitness = fitness_scores.max()
            best_individual = population[fitness_scores.argmax()].copy()
        
        sorted_indices = np.argsort(fitness_scores)[::-1]
        population = population[sorted_indices[:population_size // 2]]
        
        new_population = []
        while len(new_population) < population_size - len(population):
            parent1, parent2 = population[np.random.choice(len(population), 2, replace=False)]
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = np.vstack([population, new_population])
        print(f"Generation {generation + 1}/{n_generations}, Best fitness: {best_fitness:.4f}")
    
    selected_features = np.where(best_individual == 1)[0]
    print(f"Selected {len(selected_features)} features: {X.columns[selected_features].tolist()}")
    
    
    return selected_features

# 3. Preprocess data with selected features
def preprocess_data(X, y, selected_features):
    X_selected = X.iloc[:, selected_features].values
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_selected)
    
    # Reshape for CNN-LSTM: (samples, timesteps, channels)
    X_scaled = X_scaled[..., np.newaxis]
    
    return X_scaled, y, scaler

# 4. Split data
def split_data(X, y, test_size=0.3):
    return train_test_split(X, y, test_size=test_size, random_state=42, stratify=y)

# 5. Build CNN-LSTM-Attention model
def build_cnn_lstm_attention_model(num_features):
    inputs = Input(shape=(num_features, 1))
    
    # CNN layer
    x = Conv1D(filters=32, kernel_size=3, activation='relu', padding='same')(inputs)
    x = MaxPooling1D(pool_size=2)(x)
    
    # LSTM layer
    x = LSTM(32, return_sequences=True)(x)
    
    # Attention layer
    attention = MultiHeadAttention(num_heads=2, key_dim=16)(x, x)
    x = LayerNormalization(epsilon=1e-6)(x + attention)
    
    # Global pooling
    x = tf.keras.layers.GlobalMaxPooling1D()(x)
    
    # Dense layers
    x = Dense(16, activation='relu')(x)
    x = Dropout(0.3)(x)
    outputs = Dense(1, activation='sigmoid')(x)
    
    model = Model(inputs=inputs, outputs=outputs)
    
    def focal_loss(gamma=2.0, alpha=0.25):
        def focal_loss_fn(y_true, y_pred):
            y_true = tf.cast(y_true, tf.float32)
            y_pred = tf.clip_by_value(y_pred, 1e-7, 1.0 - 1e-7)
            pt = tf.where(tf.equal(y_true, 1), y_pred, 1 - y_pred)
            loss = -alpha * tf.pow(1.0 - pt, gamma) * tf.math.log(pt)
            return tf.reduce_mean(loss)
        return focal_loss_fn
    
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
                  loss=focal_loss(),
                  metrics=['accuracy'])
    return model

# 6. Evaluate and visualize performance
def evaluate_and_visualize(y_true, y_pred, history):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    
    print("\nModel Performance Metrics:")
    print(f"Accuracy:  {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1 Score:  {f1:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=['Normal', 'Attack']))
    
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
    plt.title('Model Accuracy over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss', marker='o')
    plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
    plt.title('Model Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal', 'Attack'], yticklabels=['Normal', 'Attack'])
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# Main execution
def main():
    print("Loading dataset...")
    X, y = load_data()
    
    # Check if selected features exist
    features_path = '/kaggle/working/selected_features.npy'
    if os.path.exists(features_path):
        print("Loading pre-selected features...")
        selected_features = np.load(features_path)
        print(f"Selected {len(selected_features)} features: {X.columns[selected_features].tolist()}")
    else:
        print("Running feature selection with Genetic Algorithm...")
        selected_features = genetic_algorithm_feature_selection(X, y)
    
    print("Preprocessing data with selected features...")
    X_scaled, y, _ = preprocess_data(X, y, selected_features)
    
    print("Splitting data...")
    X_train, X_test, y_train, y_test = split_data(X_scaled, y)
    
    print("Building CNN-LSTM-Attention model...")
    model = build_cnn_lstm_attention_model(num_features=X_train.shape[1])
    
    class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
    class_weight_dict = {0: class_weights[0], 1: class_weights[1]}
    print(f"Class weights: {class_weight_dict}")
    
    print("Training model...")
    start_time = time.time()
    history = model.fit(
        X_train, y_train,
        epochs=40,
        batch_size=64,
        validation_split=0.2,
        callbacks=[
            ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=1, min_lr=0.0001)
        ],
        class_weight=class_weight_dict,
        verbose=1
    )
    end_time = time.time()
    
    training_time = end_time - start_time
    minutes = int(training_time // 60)
    seconds = int(training_time % 60)
    print(f"Training time: {minutes} minutes {seconds} seconds")
    
    print("\nModel Parameters:")
    model.summary()
    total_params = model.count_params()
    trainable_params = np.sum([np.prod(v.shape) for v in model.trainable_weights])
    non_trainable_params = total_params - trainable_params
    print(f"- Total parameters: {total_params:,}")
    print(f"- Trainable parameters: {trainable_params:,}")
    print(f"- Non-trainable parameters: {non_trainable_params:,}")
    
    print("\nMaking predictions...")
    y_pred_proba = model.predict(X_test, batch_size=64, verbose=0)
    y_pred = (y_pred_proba > 0.5).astype(int).flatten()
    
    print("Evaluating and visualizing results...")
    evaluate_and_visualize(y_test, y_pred, history)
    

if __name__ == "__main__":
    main()
```
